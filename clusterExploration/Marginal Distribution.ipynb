```python
filtered_columns = [col for col in all_columns if col.startswith('intensity')]
```

```python
data = data.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[filtered_columns] = scaler.fit_transform(data[filtered_columns])
```

```python
import scipy.cluster.hierarchy as shc
data = data.iloc[0:30000]
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
data = data.iloc[0:30000]
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(
    n_clusters=3, affinity='euclidean', linkage='ward')

cluster.fit(data.values)
labels = cluster.labels_
labels
```

```python
import matplotlib.pyplot as plt

data_array = data.values

for feature in range(data_array.shape[1]):
    plt.figure(figsize=(10, 6))
    for cluster_num in np.unique(labels):
        plt.hist(data_array[labels == cluster_num, feature], bins=20, alpha=0.5, label=f'Cluster {cluster_num}')

    plt.xlabel(f'Feature {feature} Value')
    plt.ylabel('Frequency')
    plt.title(f'Feature {feature} Distribution by Cluster')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
data['Cluster_Labels'] = labels
```

```python
data
```

```python
import numpy as np
from scipy.stats import f_oneway
p_values = {'Cluster 0': {}, 'Cluster 1': {}, 'Cluster 2': {}}

for feature_name in data.columns[:-1]:
    cluster_groups = [data[feature_name][data['Cluster_Labels'] == i] for i in range(3)]
    f_statistic, p_value = f_oneway(*cluster_groups)
    for i, cluster_name in enumerate(p_values.keys()):
        p_values[cluster_name][feature_name] = p_value
for cluster_name, feature_p_values in p_values.items():
    print(f"P-values for {cluster_name}:")
    for feature, p_value in feature_p_values.items():
        print(f"{feature}: {p_value}")
```

```python
from sklearn.decomposition import PCA

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster_labels = cluster.fit_predict(data.values)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(data.values)

pca_df = pd.DataFrame({'PCA1': pca_result[:, 0], 'PCA2': pca_result[:, 1], 'Cluster': cluster_labels})

plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['Cluster'])
plt.title('2D PCA Visualization of Clusters')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Cluster')
plt.show()
```
