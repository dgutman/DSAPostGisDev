```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
    df = pd.read_csv("MAP01938_0000_0E_01_region_001_quantification.csv")
```

```python
df
```

```python
all_columns = df.columns
```

```python
filtered_columns = [col for col in all_columns if col.startswith('intensity')]
```

```python
data = df[filtered_columns]
```

```python
data = data.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[filtered_columns] = scaler.fit_transform(data[filtered_columns])
```

```python
data.head()
```

```python
correlation_matrix = data.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
mod_data = df[filtered_columns]
mean_values = np.mean(mod_data , axis=0)
std_dev_values = np.std(mod_data , axis=0)
min_values = np.min(mod_data , axis=0)
max_values = np.max(mod_data , axis=0)
```

```python
for i in range(mod_data.shape[1]):
    print(f"Feature {i + 1}:")
    print(f"  Mean: {mean_values[i]:.2f}")
    print(f"  Standard Deviation: {std_dev_values[i]:.2f}")
    print(f"  Min: {min_values[i]:.2f}")
    print(f"  Max: {max_values[i]:.2f}")
    print("----------------------")
```

```python
columns_to_drop = ['intensity_mean_CD11B', 'intensity_mean_CD45B', 'intensity_mean_FOXP3','intensity_mean_PDL1']
data = data.drop(columns_to_drop, axis=1)
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
data = data.iloc[0:30000]
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(
    n_clusters=3, affinity='euclidean', linkage='ward')

cluster.fit(data.values)
labels = cluster.labels_
labels
```

```python
from sklearn.decomposition import PCA

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster_labels = cluster.fit_predict(data.values)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(data.values)

pca_df = pd.DataFrame({'PCA1': pca_result[:, 0], 'PCA2': pca_result[:, 1], 'Cluster': cluster_labels})

plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['Cluster'])
plt.title('2D PCA Visualization of Clusters')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Cluster')
plt.show()
```
